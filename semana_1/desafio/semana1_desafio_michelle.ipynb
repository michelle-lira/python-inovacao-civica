### Importe as bibliotecas que usará no código

### Determine sua idade

Para calcular a idade vou começar verificando o ano atual e armazenando em uma variável:

```python
ano_atual = dt.today().year
#print(ano_atual)

# Em seguida armazenaremos a idade da usuária:
ano_nascimento = int(input("Qual o seu ano de seu nascimento? (YYYY) "))
```

```python
# Calculando a idade atual da usuária
idade_atual = ano_atual - ano_nascimento
idade_atual
```

### Leia o csv e converta para dicionário

```python
# Vou dar uma olhada no arquivo primeiro usando pandas
candidaturas = pd.read_csv("consulta_cand_2020_PE_desafio.csv", sep=',', encoding='latin-1')
candidaturas.head()
```

```python
candidaturas.shape
```

```python
lista_dados = []

with open('consulta_cand_2020_PE_desafio.csv', 'r', encoding='latin-1') as candidaturas:
    csv = reader(candidaturas)
    lista_dados = list(csv)
    print(lista_dados)
```

###  Quantas pessoas candidatas no estado de Pernambuco seriam mais novas do que você 

```python
# Removendo os títulos (linha de índice 0), das colunas do csv e armazenando na variável chaves
chaves = lista_dados.pop(0)
print(chaves)
```

**Para identificar a idade de cada candidato, vamos percorrer lista por lista do arquivo e       
depois transformar cada lista em um dicionário:
a chave será o título da coluna e terá o seu valor correspondente.**

```python
lista_dicionarios = []

for candidaturas in lista_dados:
    itens_zip = zip(chaves, candidaturas) # zip(): cria listas de tuplas usando as listas "chaves" e "candidaturas"
    dicionario = dict(itens_zip) # converte as listas de tuplas em dicionários
    lista_dicionarios.append(dicionario)
```

```python
print(lista_dicionarios[:5])
```

```python
# Checando os candidatos mais novos

# Armazenando a lista de diconários na variável "total_candidaturas"
total_candidaturas = len(lista_dicionarios)

# Estabelecendo um contador
candidaturas_mais_jovens = 0

for candidatura in lista_dicionarios:
    if int(candidatura['NR_IDADE_DATA_POSSE']) < idade_atual:
        candidaturas_mais_jovens += 1
        
print(candidaturas_mais_jovens)
```

### Quantos porcentos isso representa em relação ao total de pessoas candidatas.

```python
percentual_candidaturas = (candidaturas_mais_jovens * 100) / total_candidaturas

percentual = round(percentual_candidaturas, 2)
#print(percentual)
```

```python
print(f'Em {ano_atual - 1} houve um total de {candidaturas_mais_jovens} candidaturas de pessoas mais jovens que eu e isto representa {percentual}% do total de candidaturas registradas.')
```
